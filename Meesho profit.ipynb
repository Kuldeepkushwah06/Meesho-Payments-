#  Total Profit for Entire Month

```python
import pandas as pd

excel_file='Meesho_Payment_Sheet_Oct1_Oct31.xlsx'

df_sheet1=pd.read_excel(excel_file,sheet_name='Order Payments')
df_sheet2=pd.read_excel(excel_file,sheet_name='Sheet1')
df_sheet3=pd.read_excel(excel_file,sheet_name='Purchase Price')

purchase_sum=df_sheet1['Purchase Price'].sum()
delivered_final_settelment=df_sheet1['Final Settlement Amount'].sum()
RTO_sheet_final_settelment=df_sheet2['Final Settlement Amount'].sum()
gst_on_shipping=df_sheet1['GST on Shipping Charge'].sum()

final_profit=(delivered_final_settelment-purchase_sum+RTO_sheet_final_settelment)
net_profit=(final_profit+gst_on_shipping)

print(f'Total profit: {round(net_profit,2)}')
```

The total profit for the entire month is Rs.13329.52

It is net profit of products with status delivered, shipped returned and RTO. 

It also contains GST charged on shipping. 

It is calculated by substracting final settelment price of delivered products from purchase price and adding the rest status product's final settelment price along  with GST charged on shipping of delivered products.

#  Profitability SKU wise for Entire Month

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

excel_file = 'Meesho_Payment_Sheet_Oct1_Oct31.xlsx'

df_sheet1 = pd.read_excel(excel_file, sheet_name='Order Payments')
df_sheet2 = pd.read_excel(excel_file, sheet_name='Sheet1')

d = {}

for index, row in df_sheet1.iterrows():
    key = row['SKU']
    
    final_settlement_sheet2 = df_sheet2.loc[df_sheet2['SKU'] == key, 'Final Settlement Amount'].values
    
    if len(final_settlement_sheet2) > 0:
        final_settlement_sheet2 = final_settlement_sheet2[0]
    else:
        final_settlement_sheet2 = 0
    
    value = row['Final Settlement Amount'] - row['Purchase Price'] + final_settlement_sheet2

    if key in d:
        d[key] += value
    else:
        d[key] = value

# Print SKU's and their Net profit
for key, value in d.items():
    print(f' {key}, Net Profit: {round(value,2)}')


# Convert the dictionary to a DataFrame
df_plot = pd.DataFrame(list(d.items()), columns=['SKU', 'Net Profit'])

# Plotting the graph
plt.figure(figsize=(12, 8))
sns.barplot(x='Net Profit', y='SKU', data=df_plot, palette='viridis')
plt.xlabel('Net Profit')
plt.ylabel('SKU')
plt.title('Net Profit by SKU')
plt.show()
```

The above bar graph shows the profitability SKU wise for entire month-

Some SKU's has positive net profit and some has negative net profit.

The SKU named  ART-MARK-ALC-12SK has highet Positive net profit i.e in profit of Rs.6722.19

The SKU named EPOXY-PRIME-2:1-750G has highet negative net profit i.e in loss of Rs.-9993.51



# Total Payment Received for Entire Month

```python
import pandas as pd

excel_file = 'Meesho_Payment_Sheet_Oct1_Oct31.xlsx'

df = pd.read_excel(excel_file)

# Columns to sum for total payment received
payment_columns = ['Final Settlement Amount','GST on Shipping Charge']

# Calculate the total payment received
total_payment_received = df[payment_columns].sum().sum()

print(f'Total Payment Received: {round(total_payment_received,2)}')
```

The Total payment Received for entire month is Rs.65595.09

It is total final settelment amount of delivered prducts and the GST chaged for shipping which will be transfered to users's account.

# Total Sales % State Wise

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

excel_file = 'Meesho_Payment_Sheet_Oct1_Oct31.xlsx'

df_sheet1 = pd.read_excel(excel_file, sheet_name='Order Payments')

d = {}
total_sales_price = df_sheet1['Total Sale Amount (Incl. Commission & GST)'].sum()

for index, row in df_sheet1.iterrows():
    key = row['SKU']
    value = row['Total Sale Amount (Incl. Commission & GST)']

    if key in d:
        d[key] += value
    else:
        d[key] = value

# Calculate sales percentages
sales_percentages = {key: (value / total_sales_price) * 100 for key, value in d.items()}

# Print SKU's and their sales % profit
for key, value in sales_percentages.items():
    print(f'{key}, sales%: {round(value, 2)}%')

# Convert the dictionary to a DataFrame
df_plot = pd.DataFrame(list(sales_percentages.items()), columns=['SKU', 'Sales Percentage'])

# Plotting the graph
plt.figure(figsize=(12, 8))
sns.barplot(x='Sales Percentage', y='SKU', data=df_plot, palette='viridis')
plt.xlabel('Sales%')
plt.ylabel('SKU')
plt.title('Total Sales Percentage')
plt.show()
```

The above graph shows the Sales percentage state wise for entire month- 

From the graph we analyse:

EPOXY-PRIME-2:1-750G has the highest sales percentage of 27.99%

papercliplondon and toolsartcraft5pcsemboss01 has lowest and equal sales percentage of 0.11%
